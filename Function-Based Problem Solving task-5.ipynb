# Problem 1: Social Media – Post Engagement Analyzer
## Problem Statement
Create a function to analyze engagement on a social media post.


```python
# creating the function
def engage_analyzer (likes):
    total = 0
    
    # calculating the total nuumber of likes
    for like in likes:
        total += like

    # checking the post status
    if total >= 1000:
        status = "Viral Post"
    else:
        status = "Normal Engagement"
        
    print("Total Likes:",total)
    print("Post Status:",status)

engage_analyzer([200, 400, 700])
```

# Problem 3: Agriculture – Rainfall Adequacy Checker
## Problem Statement
Create a function that analyzes seasonal rainfall data.


```python
def rainfall_check(data, required):
    avg = sum(data)/len(data)

    if avg >= required:
        status = "Adequate Rainfall"
    else:
        status = "Inadequate Rainfall"

    print("Avarage Rainfall:",avg)
    print("Rainfall Status:",status)

rainfall_check([78, 57, 98, 70, 80], 70)
```

# Problem 4: Social Media – Duplicate Account Detection
## Problem Statement
Create a function to detect duplicate usernames in a system.


```python
def check_duplicate(users):
    if len(users) != len(set(users)):
        status = "Yes"
    else:
        status = "No"

    print("Duplicate Accounts Found:",status)

check_duplicate(["abhi", "babu", "abhi", "ramu"])
```

# Problem 6: Agriculture – Premium Crop Price Filter
## Problem Statement
Create a function that filters crop prices above a premium threshold.


```python
def crop_premium(prices):
    pre = []
    for price in prices:
        if price > 2000:
            pre.append(price)
    print("Premium Crops:",pre)

crop_premium([1500, 2500, 1900, 3200, 1200])
```

# Problem 7: System Monitoring – Application Health Checker

# Problem 9: E-Learning – Student Attendance Eligibility System
## Problem Statement

```python
def attendance_check(records):
    percent = (sum(records) / len(records)) * 100
    if percent >= 75:
        status = "Eligible"
    else:
        status = "Not Eligible"

    print("Attendance Percentage:",percent)
    print("Exam Eligibility:",status)

attendance_check([1,1,1,0,1,1,1,1,0,1])
```

# Problem 10: Smart Electricity Bill Analyzer
## Problem Statement
You are building a smart electricity billing system for residential users.
Create a function that calculates the electricity bill and classifies usage level based on units consumed.


```python
def electricity_bill(units):
    bill = 0

    if units <= 100:
        bill = units * 3
    elif units <= 200:
        bill = (100 * 3) + (units - 100) * 5
    else:
        bill = (100 * 3) + (100 * 5) + (units - 200) * 7

    if bill < 500:
        status = "Low Usage"
    elif bill <= 1500:
        status = "Moderate Usage"
    else:
        status = "High Usage"

    return bill, status

amount, level = electricity_bill(250)
print("Bill:",amount)
print("Usage:",level)
```
