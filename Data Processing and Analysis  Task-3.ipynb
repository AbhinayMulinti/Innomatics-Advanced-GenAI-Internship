# Problem Statement 1: Employee Performance Bonus Eligibility

```python
# Input in dictionary format
employees = {"Ravi" : 92, "Anita" : 88, "Kiran" : 92, "Suresh" : 85}

# Finding highst score 
high_score = max(employees.values())

# Get name of employees with the highest score
top_perform = []
for name, score in employees.items():
    if score == high_score:      
        top_perform.append(name)

# Printing the result 
print("Top Performers Eligible for Bonus: ", " , ".join(top_perform), "(Score: ", high_score, ")")
```

# Problem Statement 2: Search Query Keyword Analysis

```python
#Input in string format
query = "Buy mobile phone buy phone online"

# Converting the string to lowercase
query = query.lower()

# Splitting the query into words 
words = query.split()

# Counting frequency of each words
word_count = {}
for word in words:
    if word in word_count:
        word_count[word] += 1
    else:
        word_count[word] = 1

# finding the word occuring more than one time
keywords = {}
for word, count in word_count.items():
    if count >= 2:
        keywords[word] = count

# printing the output
print(keywords)
```

# Problem Statement 3: Sensor Data Validation


```python
# input in list format
sensor_readings = [3, 4, 7, 8, 10, 12, 5]

# empty list to store readings
valid_readings = []

for i, j in enumerate(sensor_readings):
    if j % 2 == 0:  # Check if even
        valid_readings.append((i,j))

# PRinting the final results
print("Valid Sensor Readings (Hour, Value):\n",valid_readings)
```

# Problem Statement 4: Email Domain Usage Analysis


```python
# Input in list format
emails = ["ravi@gmail.com", "anita@yahoo.com", "kiran@gmail.com", "suresh@gmail.com", "meena@yahoo.com"]

# for storing each domain count
domain_count = {}

# domains counting
for i in emails:
    domain = i.split("@") [1]
    if domain in domain_count:
        domain_count[domain] += 1
    else:
        domain_count[domain] = 1

# percantage calculation and printing output
for i, j in domain_count.items():
    print(f"{i}: {(j/len(emails))*100:.0f}%")
```

# Problem Statement 5: Sales Spike Detection


```python
# given sales list 
sales = [1200, 1500, 900, 2200, 1400, 3000]

# Calculating the average
avg_sales = sum(sales) / len(sales)

# More than 30% above average
threshold = avg_sales + (avg_sales * 0.3)

# Printing the output
for i, j in enumerate(sales, start=1):
    if j >= threshold:
        print(f"Day {i}: {j}")
```

# Problem Statement 6: Duplicate User ID Detection


```python
# Input in the list format
user_ids = ["user1", "user2", "user1", "user3", "user1", "user3"]

# Counting the duplicates
count_ids = {}
for user in user_ids:
    if user in count_ids:
        count_ids[user] += 1
    else:
        count_ids[user] = 1

# printing the output
for i, j in count_ids.items():
    if j >= 2:
        print(i,"->",j,"times")
```
