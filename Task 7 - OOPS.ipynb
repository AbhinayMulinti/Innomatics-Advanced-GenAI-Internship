# Problem 1: Smart Light Controller
## Problem Statement
Create a class to control a smart light that can be turned ON or OFF and display its status.


```python
class smart_light:
    def __init__(self, light_name, light_status):
        self.name = light_name
        self.status = light_status

    def turn_on(self):
        if self.status == "ON":
            print(f"{self.name} is {self.status}")
        else:
            print(f"{self.name} is {self.status}")

light = smart_light("Bedroom Light", "ON")
light.turn_on()
```

# Problem 2: Employee ID Card System
## Problem Statement
Create a class that stores employee name, ID, and department and displays ID card details.


```python
class employee_details:
    def __init__(self, emp_name, emp_id, emp_dep):
        self.emp_name = emp_name
        self.emp_id = emp_id
        self.emp_dep = emp_dep

    def display_id_card(self):
        name = self.emp_name.title()
        Id = self.emp_id.upper()
        Dep = self.emp_dep.upper()
        print("Employee ID Card")
        print("Name:",name)
        print("ID:",Id)
        print("Department:",Dep)

id_card = employee_details("rahul", "emp102", "it")
id_card.display_id_card()
```

# Problem 3: Mobile Contact Record
## Problem Statement
Create a class to store contact name and phone number and display contact information.


```python
class mobile_contact:
    def __init__(self, name, contact_num):
        self.name = name
        self.number = contact_num

    def display_mobile_number(self):
        if (len(self.number) == 10) & (self.number.isdigit()):
            print("Contact Saved")
            print("Name:",self.name.title())
            print("Phone:",self.number)
        else:
            print("Enter the correct mobile number")

contact = mobile_contact("anita", "9876543210")
contact.display_mobile_number()
```

# Problem 4: Product Price Tag Generator
## Problem Statement
Create a class that stores product name and price and prints a formatted price tag.


```python
class product_price:
    def __init__(self, product_name, product_price):
        self.name = product_name
        self.price = product_price

    def display(self):
        print("Product:",self.name.title())
        print(f"Price: ₹{self.price}")

tag = product_price("headphones", 2499)
tag.display()
```

# Problem 5: Movie Rating Display System
## Problem Statement
Create a class that stores movie name and rating and displays movie details.


```python
class movie_rating:
    def __init__(self, movie_name, movie_rating):
        self.name = movie_name
        self.rating = movie_rating

    def display(self):
        print("Movie:",self.name.title())
        print(f"Price:",(self.rating),"/ 5")

movie = movie_rating("inception", 4.8)
movie.display()
```

# Problem 6: Delivery Address Manager
## Problem Statement
Create a class that stores customer name and delivery address and prints delivery details.


```python
class delivery_address:
    def __init__(self, cus_name, cus_address):
        self.name = cus_name
        self.address = cus_address

    def display(self):
        name = self.name.title()
        add = self.address.title()
        print("Delivery Details")
        print("Name:",name)
        print("Address:",add)

manager = delivery_address("suman", "hyderabad")
manager.display()
```
