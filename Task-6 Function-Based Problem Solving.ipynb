# Problem Statement 1:Smart Parking Lot Management System
## Design a function to manage a smart parking lot.

```python
def smart_parking(capacity, logs):
    parked = 0

    for park in logs:
        park = park.upper()
        if park == "IN":
            parked += 1
        elif park == "OUT":
            parked -= 1

    print("Currently Parked Vehicles:",parked)

    if parked >= capacity:
        status = "Exceeded Capacity"
    else:
        status = "Available"

    print("Parking Status:",status)

smart_parking(50, ["IN", "IN", "in", "OUT", "IN", "in", "out"])
```

# Problem Statement 2:Online Food Delivery Time Estimator
## Create a function that estimates delivery time based.


```python
def food_delivery_time(distance, traffic, weather):
    base_time = distance * 4
    traffic = traffic.capitalize()
    weather = weather.capitalize()

    if traffic == "High":
        base_time += 15
    elif traffic == "Medium":
        base_time += 10
    else:
        base_time += 5

    if weather == "Rainy":
        base_time += 8
    elif weather == "Storm":
        base_time += 20

    print("Estimated Delivery Time:",base_time)

food_delivery_time(8, "high", "Rainy")
```

# Problem Statement 3:Movie Theatre Seat Occupancy Analyzer
## Build a function that analyzes seat booking data 

```python
def theatre_seat_occupancy(total_seats, booked_seats):
    count = len(booked_seats)
    occupancy = (count / total_seats) * 100
    
    if occupancy == 100:
        status = "Housefull"
    elif 75 <= occupancy < 100:
        status = "Almost Full"
    else:
        status = "Seats Available"

    print(f"Occupancy: {int(occupancy)}%")
    print("Show Status:",status)

theatre_seat_occupancy(200, [1]*150)
```

# Problem Statement 4:Cloud Server Load Classification System
## Create a function to classify server load based on CPU usage readings.

```python
def cloud_server_load(cpu_readings):
    total = 0

    for reading in cpu_readings:
        total += reading

    average = total / len(cpu_readings)

    if average < 50:
        status = "Normal"
    elif 50 <= average <= 80:
        status = "Warning"
    else:
        status = "Critical"

    print(f"Average CPU Load: {int(average)}%")
    print("Server Status:",status)

cloud_server_load([45, 60, 70, 85, 90])
```

# Problem Statement 5:Smart Classroom Resource Usage Monitor
## Design a function that tracks usage of classroom resources (projector, AC, lights) and identifies overuse patterns.


```python
def smart_classroom_resource(usage):
    overused = []

    for resource in usage:
        if usage[resource] > 8:
            overused.append(resource)

    if overused:
        print("Overusedd Resource:",",".join(overused))
        print("Energy Alert: Yes")
    else:
        print("Overusedd Resource: None")
        print("Energy Alert: No")


smart_classroom_resource({"Projector":6, "AC":9, "Lights":4})
```

# Problem Statement 6: Online Event Registration Capacity Controller
## Create a function that manages event registrations 


```python
def online_event_registration(capacity, registers):
    if registers > capacity:
        confirmed = capacity
        waitlist = registers - capacity
        status = "Closed"
    else:
        confirmed = registers
        waitlist = 0
        status = "Open"

    print("Confirmed Registrations:",confirmed)
    print("Waitlisted Users:",waitlist)
    print("Registrations status:",status)

online_event_registration(100, 105)
```
