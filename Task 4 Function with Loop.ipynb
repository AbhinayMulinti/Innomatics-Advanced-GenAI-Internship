# Problem 1: Mobile Recharge Validation System


# Problem Statement
Create a function to validate a mobile recharge request.
## Rules:
Recharge amount must be ≥ ₹50

Valid plans: 199, 299, 399, 599

User should be allowed to retry if invalid




```python
# creating the function with one parameter
def mobile_recharge(user_plan):
    recharge_plans = [199, 299, 399, 599]
    
    # entering while loop
    while True:
        
        # checking the plan is valid or not
        if user_plan in recharge_plans:
            return "Recharge is valid"
            break
        else:
            user_plan = int(input("Enter the valid recharge plan : "))
```

```python
# calling the function
mobile_recharge(99)
```

# Problem 2: Inventory Reorder Alert System


# Problem Statement
## Given a dictionary of products and stock quantities:
Print Reorder Alert if stock < 15

Print Stock OK otherwise

```python
# creating the function with parameters
def stock_quantity(products):
    
    # checking the quantity of each product
    for item, quantity in products.items():
        if quantity < 15:
            print(item,"--> Reorder Alert\n")
        else:
            print(item, "--> Stock OK\n")
```

```python
# calling the function with passing dictionary
stock_quantity({"Apples":20, "Carrot":14, "Pen":23, "pencil":5, "books":10})
```

# Problem 3: Student Result Processing System

# Problem Statement
## Create a function that:
Accepts a list of student marks

Calculates average marks

Prints Pass if average ≥ 50 else Fail


```python
# creating the function with one parameter
def student_grade(marks):
    
    # finding the average of the student
    average = sum(marks) / len(marks)
    
    # checking student is pass or fail
    if average >= 50:
        print("The student is pass")
    else:
        print("The student is fail")
```

```python
# calling the function by passing marks in list format
student_grade([60, 97, 52, 40, 83, 99, 83])
```

# Problem 4: Cab Fare Estimator with Retry Option


# Problem Statement
## Create a cab fare calculator:
Base fare = ₹50

₹12 per km

Peak hour → 25% extra

Allow retry using while loop

```python
# creating the function with no parameters
def cab_fare():
    # Asking the distance and peak hour or not
    distance = int(input("Enter the distance in KM :"))
    peak = input("Is it peak hour? (yes/no): ").lower()

    # calculating the fare amount
    fare_amount = 50 + (distance * 12)

    while True:
        
        # At peak hour
        if peak == "yes":
            total_amount = fare_amount + (fare_amount * 0.25)
            print(f"Total Fare = {total_amount}Rs.")
        else:
            print(f"Total Fare = {fare_amount}Rs.")

        # Asking the customer about booking
        asking = input("You want to book cab? (yes/no) :").lower()

        # booking confirm
        if asking == "yes":
            print("Booking confirmed!!")
            print("Thank you and have a safe ride")
            break
        else:
            print("Booking canclled!!")

        # asking to retry
        retry = input("You want to retry? (yes/no) :")
        if retry == "no":
            break
```

```python
cab_fare()
```

# Problem 5: Employee Attendance Eligibility Checker


# Problem Statement
## Create a function that:
Accepts attendance list ("P", "A")

Calculates attendance percentage

Returns Eligible if ≥ 75% else Not Eligible

```python
# creating the function with one parameter
def students_eligibility(Attendance):

    # counting how many presents
    present_count = 0
    for present in Attendance:

        # converting into upper case letter
        present = present.upper()
        if present == "P":
            present_count += 1

    # calculating the percentage of present
    percentage = (present_count / len(Attendance)) * 100

    # checking eligibility
    if percentage >= 75:
        return "Eligible"
    else:
        return "Not Eligible"
```

```python
students_eligibility(["P", "P", "A", "p", "p", "p", "p","a", "p", "p"])
```

# Problem 6: Password Strength Checker


# Problem Statement
Create a function to check password strength.
## Rules:
Minimum length = 8

Must contain at least one digit

Must contain at least one special character (@#$)


```python
# creating the function
def password_checker():
    password = input("Enter the password : ")
    upper = 0
    lower = 0
    digit = 0
    special = 0

    # checking all the characters
    for ch in password:
        if ch.isupper():
            upper += 1
        if ch.islower():
            lower += 1
        if ch.isdigit():
            digit += 1
        if ch in "@#$":
            special += 1

    # checking the strength of the password
    if len(password) >= 8 and upper >= 1 and lower >= 1 and digit >= 1 and special >= 1:
        return "Strong Password"
    else:
        return "Weak Password"
```

```python
password_checker()
```
