# Problem 1: Unique Words in a Sentence
## Problem Statement
- Given a sentence, identify all unique words and display:
- The count of unique words
- The unique words themselves

```python
statement = input("Enter the sentence : ")
a = statement.split()
b = set(a)
print("Unique words count is :",len(b))
print("Unique words :",b)
```

# Problem 2: Highest Salary from Employee Data
## Problem Statement
- Store employee names and their salaries, and determine which employee has the highest salary.

```python
salary = {"Ravi":75000,"Anita":68000,"Kiran":72000}
highest = 0
employee = ""
for name in salary:
    if salary[name] > highest:
        employee = name
        highest = salary[name]
print(employee,":",highest)
```

# Problem 3: Find Maximum and Minimum Values
## Problem Statement
- Find the maximum and minimum values from a list of numbers.


```python
numbers = [45, 22, 89, 10, 66]
high = numbers[0]
low = numbers[0]
for num in numbers:
    if num > high:
        high = num
    if num < low:
        low = num
print(f"Max:{high}")
print(f"Min:{low}")
```

# Problem 4: Count Products Above a Price Threshold
## Problem Statement
- Count how many products have a price greater than 1000.


```python
prices = [450, 1200, 899, 1500, 300]
count = 0
for value in prices:
    if value > 1000:
        count += 1
print("Products above 1000:",count)
```

# Problem 5: Calculate Attendance Percentage
## Problem Statement
- Calculate the attendance percentage from a list containing "P" (Present) and "A" (Absent).


```python
attendance = ["P", "p", "A", "P", "p"]
pres_count = 0
tot_count = 0
for i in attendance:
    tot_count += 1
    i = i.upper()
    if i == "P":
        pres_count += 1
print("Attendance Percentage:",(pres_count / tot_count) * 100)
```

# Problem 7: Count Character Frequency
## Problem Statement
- Count how many times each character appears in a given string.


```python
statement = input("Enter the statement :")
count = {}
for ch in statement:
    ch = ch.lower()
    if ch != " ":
        if ch in count:
            count[ch] += 1
        else:
            count[ch] = 1
print(count)
```

# Problem 10: Calculate Average Marks
## Problem Statement
- Calculate the average marks of a student from a list of marks.


```python
marks = [87, 65, 77, 49, 95, 92, 74, 53]
total = 0
for num in marks:
    total += num
print("Average Marks:",total/len(marks))
```
